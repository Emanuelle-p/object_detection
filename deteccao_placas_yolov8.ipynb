```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt

def exibir_img(caminho):
  w = 15
  h = 15
  fig = plt.figure(figsize=(8, 8))
  columns = 4
  rows = 5
  caminho_img = glob.glob(f'{caminho}/*jpg')

  quant = 0
  imgs = []
  for item in caminho_img:
    if quant <70:
      img = cv2.imread(item)
      imgs.append(img)
      quant += 1

  for i in range(1, columns*rows +1):
      fig.add_subplot(rows, columns, i)
      plt.axis("off")
      plt.imshow(cv2.cvtColor(imgs[i-1], cv2.COLOR_BGR2RGB))
  return plt.show()
```

```python
caminho_imagens = '/content/runs/detect/predict2'

exibir_img(caminho_imagens)
```
